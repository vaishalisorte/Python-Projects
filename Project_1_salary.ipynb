Data **visulazations**

```python
plt.figure(figsize=(5,4))
sns.barplot(x='sex',y='salary',data=data,color='green')
plt.title('salary by sex')
plt.show()
```

```python
plt.hist(data.salary,color='red',edgecolor='green', bins=5)
plt.title('salary')
plt.show()
```

```python
x = data['sex'].value_counts()
```

```python
plt.figure(figsize=(5,5))
plt.pie(x,labels=x.index,colors=['red','green'],explode=[0.1,0],shadow=True,autopct='%0.1f%%')
plt.show()
```

```python
y = data['rank'].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.pie(y,labels=y.index,shadow=True,explode=[0,0,0.1],autopct='%1.2f%%',colors=['red','green','pink'])
plt.show()
```

```python
plt.pie(x,labels=x.index,colors=['red','green'])
circle=plt.Circle((0,0),color='white',radius=0.50)
plt.gca().add_artist(circle)
plt.show()
```

```python
plt.scatter(data.salary,data.service,color='purple')
plt.xlabel('salary')
plt.ylabel('service')
plt.show()
```

```python
plt.boxplot(data.phd,notch=True)
plt.show()
```

```python
plt.plot(data.salary,data.service)
plt.show()
```

```python
sns.displot(data.service,color='pink')
plt.show()
```

```python
sns.countplot(x='sex',data=data,hue='service')
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
sns.lineplot(x='service',y='phd',data=data, markers='*')
plt.show()
```
