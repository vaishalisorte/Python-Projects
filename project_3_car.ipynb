**Data Visulalaztions**

```python
x=list(data['Name'].value_counts()[0:5].keys())
y=data['Name'].value_counts()[0:5]
```

```python
plt.figure(figsize=(12,6))
plt.bar(x,y,color=['green','red','purple','blue','pink'])
plt.show()
```

```python
sns.lineplot(x='year',y='selling_price',data=data,color='green')
plt.show()
```

```python
plt.hist(data.mileage,color='green',edgecolor='red',bins=5)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data.mileage,data.selling_price,color='green')
plt.show()
```

```python
x = data['fuel'].value_counts()
```

```python
plt.figure(figsize=(10,5))
plt.pie(x,labels=x.index,autopct='%1.2f%%')
plt.show()
```

```python
v=data['seller_type'].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.pie(v,labels=v.index,explode=[0,0,0.2],autopct='%1.2f%%',shadow=True)
circle=plt.Circle((0,0),color='white',radius=0.50)
plt.gca().add_artist(circle)
```
